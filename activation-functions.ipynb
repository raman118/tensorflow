# sigmoid

```python
import math
```

```python
def sigmoid(x):
    return 1/1+(math.exp(-x))
```

```python
sigmoid(5)
```

# tanh

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-56)
```

# relu

```python
def relu(x):
    return max(0 , x)
```

```python
relu(-6)
```

# leaky relu

```python
def leaky_relu(x):
    return max(0.1*x , x)
```

```python
leaky_relu(6)
```

```python
leaky_relu(-7)
```
